```python
# INTRODUCCION A MATPLOTLIB
# PRACTICA 1: HACIENDO UNA VISUALIZACION
# Cargar base de datos datos de clima de dos cuidades en USA: seatle vs. austin
import pandas as pd

# Reemplaza 'nombre_del_archivo.csv' con el nombre real de tu archivo
seattle_weather = pd.read_csv('/content/seattle_weather.csv')

# Muestra las primeras filas del DataFrame para verificar la importación
print(seattle_weather.head())
```

```python
# Reemplaza 'nombre_del_archivo.csv' con el nombre real de tu archivo
austin_weather = pd.read_csv('/content/austin_weather.csv')

# Muestra las primeras filas del DataFrame para verificar la importación
print(austin_weather.head())
```

```python
# Introduccion a la interfaz pyplot
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
plt.show()
```

```python
# Agregando datos los ejes SEATLE
seattle_weather["DATE"]
```

```python
# Agregando datos los ejes SEATLE
seattle_weather["MLY-TAVG-NORMAL"]
```

```python
# Agregando datos los ejes AUSTIN
austin_weather["DATE"]
```

```python
# Agregando datos los ejes AUSTIN
austin_weather["MLY-TAVG-NORMAL"]
```

```python
# Agregando datos a los ejes SEATLE
# Graficar un gráfico de líneas con las primeras 12 meses
plt.figure(figsize=(10, 6))  # Ajusta el tamaño de la figura si es necesario
plt.plot(seattle_weather['DATE'].head(12), seattle_weather['MLY-TAVG-NORMAL'].head(12))
plt.xlabel('DATE')
plt.ylabel('MLY-TAVG-NORMAL')
plt.title('Temperatura promedio mensual normal en Seattle (primeras 12 meses)')
plt.xticks(rotation=45, ha='right') # Rotar las etiquetas del eje x para mejor legibilidad
plt.tight_layout() # Ajustar el diseño para evitar que las etiquetas se superpongan
plt.show()
```

```python
# Agregando datos a los ejes AUSTIN
fig, ax = plt.subplots()
ax.plot(austin_weather["DATE"], austin_weather["MLY-TAVG-NORMAL"])
plt.show()
```

```python
# Juntando todo SEATLE va AUSTIN
fig, ax = plt.subplots()
ax.plot(seattle_weather['DATE'].head(12), seattle_weather['MLY-TAVG-NORMAL'].head(12))
ax.plot(austin_weather["DATE"], austin_weather["MLY-TAVG-NORMAL"])
plt.show()
```

```python
# Seleccionando color
fig, ax = plt.subplots()
ax.plot(seattle_weather['DATE'].head(12), seattle_weather['MLY-TAVG-NORMAL'].head(12), marker="v", linestyle="--", color="r")
plt.show()
# Revisar como usar otros paletas de color en el siguiente link:
# https://matplotlib.org/stable/gallery/color/index.html
```

```python
# Personalizando etiquetas de ejes
fig, ax = plt.subplots()
ax.plot(seattle_weather['DATE'].head(12), seattle_weather['MLY-TAVG-NORMAL'].head(12), marker="v", linestyle="--", color="r")
ax.set_xlabel("Tiempo (Meses)")
plt.show()
```

```python
# Configurando etiqueta en Eje Y
fig, ax = plt.subplots()
ax.plot(seattle_weather['DATE'].head(12), seattle_weather['MLY-TAVG-NORMAL'].head(12), marker="v", linestyle="--", color="r")
ax.set_xlabel("Tiempo (Meses)")
ax.set_ylabel("Temperatura Promedio (Fahrenheit)")
plt.show()
```

```python
# Agregando un titulo
fig, ax = plt.subplots()
ax.plot(seattle_weather['DATE'].head(12), seattle_weather['MLY-TAVG-NORMAL'].head(12), marker="v", linestyle="--", color="r")
ax.set_title("Clima en Seatle")
ax.set_xlabel("Tiempo (Meses)")
ax.set_ylabel("Temperatura Promedio (Fahrenheit)")
plt.show()
```

```python
# Agregando datos todos los años
fig, ax = plt.subplots()
ax.plot(seattle_weather['DATE'], seattle_weather['MLY-PRCP-NORMAL'], color="b")
ax.set_xlabel("Tiempo (Meses)")
ax.set_ylabel("Precipitacion (pulgadas)")
plt.show()
```

```python
# Agregando más datos solo el ultimo año
fig, ax = plt.subplots()
ax.plot(seattle_weather['DATE'].head(12), seattle_weather['MLY-PRCP-25PCTL'].head(12), linestyle="--", color="b")
ax.plot(seattle_weather['DATE'].head(12), seattle_weather['MLY-PRCP-NORMAL'].head(12), color="b")
ax.plot(seattle_weather['DATE'].head(12), seattle_weather['MLY-PRCP-75PCTL'].head(12), linestyle="--", color="b")
ax.set_xlabel("Tiempo (Meses)")
ax.set_ylabel("Precipitacion (pulgadas)")
plt.show()
```

```python
# Mucho más datos
fig, ax = plt.subplots()
ax.plot(seattle_weather['DATE'].head(12), seattle_weather['MLY-PRCP-25PCTL'].head(12), linestyle="--", color="b")
ax.plot(seattle_weather['DATE'].head(12), seattle_weather['MLY-PRCP-NORMAL'].head(12), color="b")
ax.plot(seattle_weather['DATE'].head(12), seattle_weather['MLY-PRCP-75PCTL'].head(12), linestyle="--", color="b")
ax.plot(austin_weather['DATE'], austin_weather['MLY-PRCP-25PCTL'], linestyle="--", color="r")
ax.plot(austin_weather['DATE'], austin_weather['MLY-PRCP-NORMAL'], color="r")
ax.plot(austin_weather['DATE'], austin_weather['MLY-PRCP-75PCTL'], linestyle="--", color="r")
ax.set_xlabel("Tiempo (Meses)")
ax.set_ylabel("Precipitacion (pulgadas)")
plt.show()
```

```python
# Subplots con datos
fig, ax = plt.subplots(2,1)
ax[0].plot(seattle_weather['DATE'].head(12), seattle_weather['MLY-PRCP-25PCTL'].head(12), linestyle="--", color="b")
ax[0].plot(seattle_weather['DATE'].head(12), seattle_weather['MLY-PRCP-NORMAL'].head(12), color="b")
ax[0].plot(seattle_weather['DATE'].head(12), seattle_weather['MLY-PRCP-75PCTL'].head(12), linestyle="--", color="b")
ax[1].plot(austin_weather['DATE'], austin_weather['MLY-PRCP-25PCTL'], linestyle="--", color="r")
ax[1].plot(austin_weather['DATE'], austin_weather['MLY-PRCP-NORMAL'], color="r")
ax[1].plot(austin_weather['DATE'], austin_weather['MLY-PRCP-75PCTL'], linestyle="--", color="r")
ax[1].set_xlabel("Tiempo (Meses)")
ax[0].set_ylabel("Precipitacion (pulgadas)")
ax[1].set_ylabel("Precipitacion (pulgadas)")
plt.show()
```

```python
# Subplots con datos
fig, ax = plt.subplots(2,1, sharey=True)
ax[0].plot(seattle_weather['DATE'].head(12), seattle_weather['MLY-PRCP-25PCTL'].head(12), linestyle="--", color="b")
ax[0].plot(seattle_weather['DATE'].head(12), seattle_weather['MLY-PRCP-NORMAL'].head(12), color="b")
ax[0].plot(seattle_weather['DATE'].head(12), seattle_weather['MLY-PRCP-75PCTL'].head(12), linestyle="--", color="b")
ax[1].plot(austin_weather['DATE'], austin_weather['MLY-PRCP-25PCTL'], linestyle="--", color="r")
ax[1].plot(austin_weather['DATE'], austin_weather['MLY-PRCP-NORMAL'], color="r")
ax[1].plot(austin_weather['DATE'], austin_weather['MLY-PRCP-75PCTL'], linestyle="--", color="r")
ax[1].set_xlabel("Tiempo (Meses)")
ax[0].set_ylabel("Precipitacion (pulgadas)")
ax[1].set_ylabel("Precipitacion (pulgadas)")
plt.show()
```
